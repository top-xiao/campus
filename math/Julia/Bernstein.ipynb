```julia
using Plots 
using SpecialFunctions
```

```julia
function log_binomial(n, k)
    if k < 0 || k > n
        return -Inf
    elseif k == 0 || k == n
        return 0
    else
        return loggamma(n+1) - loggamma(k+1) - loggamma(n-k+1)
    end
end
```

```julia
c=10
```

```julia
function g(x, n, c)
    y = exp(log_binomial(c, n) + n*log(x) + (c-n)*log(1-x))
    return y
end
```

```julia
function plot_g(n_values, c)
    p = plot(xlabel="x", ylabel="g(x, n)")
    x_values = 0:0.001:1
  
    for n in n_values
        plot!(p, x_values, g.(x_values, n, c), label="g(x, $n)")
    end
    
    display(p)
end

plot_g(0:1:c, c)
```
